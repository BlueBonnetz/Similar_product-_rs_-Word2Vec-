```python
# สร้าง word vectors จากคำอธิบายของสินค้า
sentences = [description.split() for description in df['description']]
model = Word2Vec(sentences, vector_size=1000, window=5, min_count=1, workers=4)

# แปลงคำอธิบายของสินค้าเป็นเวกเตอร์
def word_vectorizer(description):
    vector = []
    for word in description.split():
        vector.append(model.wv[word])
    return sum(vector) / len(vector)

# สร้างเมทริกซ์ของคำอธิบายสินค้า
vectors = [word_vectorizer(description) for description in df['description']]

# คำนวณความคล้ายคลึงระหว่างคำอธิบายของสินค้าโดยใช้ cosine similarity
similarity_matrix = cosine_similarity(vectors)

# เปลี่ยนให้ค่าในเมทริกซ์เป็นทศนิยม 2 ตำแหน่ง
similarity_matrix = np.round(similarity_matrix, 2)

#print("ความคล้ายคลึงระหว่างสินค้า:")
#print(similarity_matrix)
```
